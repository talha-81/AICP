### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
```

### Reading CSV File

```python
df = pd.read_csv('menu.csv')
df.head()
```

### Dealing With the specified Columns

```python
attributes = df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber',
                'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                'Calcium (% Daily Value)', 'Iron (% Daily Value)']]
attributes.head()
```

## Question No 4 
Draw boxplot for Calories vs Category to spot outliers and max calories category
### Answer

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Calories',y='Category',data=df,hue='Category')
plt.title('boxplot for Calories vs Category')
plt.xlabel('Calories')
plt.ylabel('Category')
plt.legend(title='Category')
plt.show()
```

# Question No 5
Figure out which exact item contains a high quantity for **['Calories', 'Total Fat', 'Carbohydrates',
'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium
(% Daily Value)', 'Iron (% Daily Value)'.**
Display summary link this:
### Answer

```python
for i in attributes:
    maximum = max(df[i])
    max_index = df[i].idxmax()
    print(i+':   '+df["Item"][max_index]+' - '+str(maximum))
```

## Question No 6
Draw Stripplot for each
category against the following
attributes **['Calories', 'Total Fat',
'Carbohydrates', 'Dietary
Fiber', 'Sugars', 'Protein',
'Vitamin A (% Daily Value)',
'Vitamin C (% Daily Value)',
'Calcium (% Daily Value)', 'Iron
(% Daily Value)'.**
### Answer

```python
for i in attributes:
    plt.figure(figsize=(8,6))
    sns.stripplot(x='Category',y=i,data=df,jitter=True,palette='Set2')
    plt.title(f"{i} Content")
    plt.xlabel("Category")
    plt.ylabel(i)
    plt.xticks(rotation=45)
    plt.show()
```

## Question No 7
Draw a horizontal bar graph for items in each **category** against the calories.
Here is one sample for all items in **Beef & Pork**, you have to do it for items in each category. Also, write
your observation.
### Answer

```python
cat =df["Category"].unique()
for i in cat:
    plt.figure(figsize=(20,20))
    cat_df = df[df['Category'] == i].sort_values(by='Calories',ascending=False)
    plt.barh(cat_df['Item'],cat_df['Calories'] )
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.title(f"Calories of Items in {i}")
    plt.gca().invert_yaxis()
    plt.show()
```
